```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def mandelbrot(xmin=-2.2, xmax=1.0, ymin=-1.5, ymax=1.5, res_x=1000, res_y=1200, max_iter=64): 
    x = np.linspace(xmin, xmax, res_x)
    y = np.linspace(ymin, ymax, res_y)
    real, imag = np.meshgrid(x, y)
    c = real + 1j * imag

    z = np.zeros_like(c)
    div_time = np.zeros(z.shape, dtype=int)

    for k in range(max_iter): 
        z = z**2 + c
        div = z*np.conj(z) > 4
        div_now = div & (div_time == 0)
        div_time[div_now] = k
        z[div] = 2

    return div_time

def plot_mandelbrot(mandelbrot, xmin=-2.2, xmax=1.0, ymin=-1.5, ymax=1.5): 
    plt.figure(figsize=(12, 12), dpi=100)
    plt.imshow(mandelbrot, cmap='twilight_shifted', extent=[xmin, xmax, ymin, ymax], origin='lower')
    plt.show() 
```

```python
mandel1 = mandelbrot(max_iter=64)
plot_mandelbrot(mandel1)
```
